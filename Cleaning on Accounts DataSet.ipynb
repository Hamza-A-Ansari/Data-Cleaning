# Cleaning on Accounts DataSet

   * Accounts DataSet is given, we have to clean it and create two new columns ('Account Number' & 'Acount Name') from this DataSet

### Import necessary module

```python
import pandas as pd
import re
```

### Load a CSV file into Pandas DataFrame

```python
data = pd.read_excel("Accounts Data.xlsx")
```

```python
header = data.loc[[7]]
```

```python
data = data[10:]
```

```python
data[["XYZ Pvt Ltd"]]
```

```python
data = data[data["XYZ Pvt Ltd"].isnull() == False]
```

```python
data
```

```python
data[["XYZ Pvt Ltd"]].isnull().value_counts()
```

```python
data = data[data['XYZ Pvt Ltd'].str.startswith('Totals') != True]
```

```python
data
```

```python
data = data[data['XYZ Pvt Ltd'].str.startswith('Net') != True]
```

```python
data
```

```python
data['XYZ Pvt Ltd']
```

```python
account_number = data["XYZ Pvt Ltd"].str.extract(r'(\d+)')
```

```python
account_number
```

```python
account_number = account_number.ffill()
```

```python
data["Account Number"] = account_number
```

```python
data[['XYZ Pvt Ltd','Unnamed: 1',"Account Number"]]
```

```python
acc_name = ''
lst = []
for index, row in data.iterrows():
    if row['XYZ Pvt Ltd'] == row['Account Number']:
        acc_name = row['Unnamed: 1']
    else:
        lst.append(acc_name)
```

```python
data = data[data['XYZ Pvt Ltd'].str.contains(r'\d+') != True]
```

```python
data
```

```python
data['Account Name'] = lst
```

```python
data = data[['Account Name','Account Number','XYZ Pvt Ltd', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4',\
        'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9',\
        'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12']]
```

```python
data
```

```python
data.columns = ['Account Name','Account Number', 'Date', 'Journal Number', 'Voucher', 'Posting Type', 'ledger Account','Description', 'Currency', 'Debit_TC', 'Credit_TC','Debit_AC', 'Credit_AC','Debit_RC', 'Credit_RC' ]
```
